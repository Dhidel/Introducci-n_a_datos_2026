#Resumen


Durante estos ejercicios se exploraron las tres formas principales de persistencia de datos en Python:

Nativa (open): Control total, ideal para archivos simples.

Pandas: Alta eficiencia para análisis de datos tabulares.

Pickle: Ideal para guardar objetos complejos de Python que no encajan en una tabla.

#Procedimiento

###Ejercicio #1

Leer archivos CSV y TXT usando context managers

```python
import csv

# Lectura de TXT y CSV
with open('datos.txt', 'r') as txt_file:
    contenido = txt_file.read()

with open('datos.csv', 'r') as csv_file:
    reader = csv.reader(csv_file)
    datos = list(reader)

# Escritura de TXT y CSV
with open('salida.txt', 'w') as txt_file:
    txt_file.write("Hola, este es un archivo de texto.")

with open('salida.csv', 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(['Nombre', 'Edad'])
    writer.writerow(['Alex', 19])
```

¿En qué parte de un Pipeline se usan?

Ingesta (Extract): Lectura de fuentes externas (archivos log, CSVs de ventas).

Procesamiento/Transformación (Transform): Lectura de datos intermedios y escritura de resultados tras la limpieza.

Carga (Load): Escritura final en el destino (un archivo listo para BI o una base de datos).

Almacenamiento (Staging): Escritura de puntos de control para no perder el progreso en procesos largos.

**¿Se podría usar un DataFrame para todos los tipos de datos (Estructurados, No
Estructurados, Semi Estructurados)?**

No exactamente.

Estructurados: Sí, es su especialidad (tablas, SQL, CSV).

Semi-estructurados: Sí (JSON, XML), aunque a veces requiere "aplanar" (flatten) los datos si están muy anidados.

No estructurados: No. Un DataFrame no es apto para almacenar directamente videos, audios o imágenes de forma eficiente. Para estos, se suelen guardar las rutas de los archivos o sus representaciones vectoriales (embeddings), pero no el contenido "bruto" como tal.

###Ejercicio #3

```python
import pickle

datos = {"usuario": "Alex", "rol": "Estudiante", "punteo": 95}

# Escribir (Serializar)
with open('data.pkl', 'wb') as f:
    pickle.dump(datos, f)

# Leer (Deserializar)
with open('data.pkl', 'rb') as f:
    datos_cargados = pickle.load(f)
```

**¿Qué tipo de dato se podría almacenar con un Pickle?**

Casi cualquier objeto de Python: listas, diccionarios, clases personalizadas, e incluso modelos de Machine Learning entrenados (como los de Scikit-Learn). Es muy útil para guardar el "estado" de una variable y recuperarla después sin tener que volver a procesar todo.

Eficiencia y Seguridad: El uso de with (context managers) evita fugas de memoria y bloqueos de archivos.

Herramienta según el dato: No hay una herramienta única; mientras que un CSV es excelente para interoperabilidad, Pickle es mejor para guardar la lógica de un objeto de Python, y Pandas para el análisis masivo.
