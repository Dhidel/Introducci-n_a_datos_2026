#Resumen de la situación actual


El programa Club Cinépolis representa una evolución del modelo tradicional de ventas hacia una arquitectura de negocio basada en datos. Antes, las cadenas de cine operaban con transacciones anónimas (solo sabían cuántos boletos se vendían), pero con la implementación del programa de fidelización, cada compra pasa a estar asociada a un perfil de usuario identificable.

Este cambio transforma completamente la lógica del negocio:

- Cada compra deja de ser solo una venta → pasa a ser un evento de datos.

- Cada usuario deja de ser anónimo → pasa a ser un activo analizable.

- Cada interacción alimenta sistemas de analítica y marketing

Desde el punto de vista técnico, el sistema funciona como un pipeline de datos distribuido, donde múltiples componentes trabajan en conjunto:

1. Capa de identidad (Okta CIC)

Gestiona autenticación y perfiles de usuario
Centraliza la identidad en app, web y POS
Reduce duplicados y fraude

2. Capa operativa (Vista Cinema / POS)

Registra transacciones en tiempo real
Conecta taquilla, dulcería y app
Actúa como punto de entrada de datos

3. Motor de reglas de lealtad

Aplica lógica de negocio:
acumulación de puntos (5% / 10%)
validación de nivel
promociones (2x1, combos, etc.)
Determina el comportamiento del sistema

4. Base de datos central

Almacena:
usuarios
transacciones
puntos
niveles
canjes
expiraciones

5. Capa analítica (Movio, BI)

Segmenta usuarios
Predice comportamiento
Permite campañas personalizadas

###Estrategia detrás del sistema

El diseño no es casual. Está optimizado para:

Incrementar frecuencia de visitas (modelo semestral por niveles)
Aumentar gasto promedio (más puntos = más consumo)
Generar datos de comportamiento (Big Data)

El sistema de niveles (Fan → Fanático → Súper Fanático) introduce un componente de gamificación, donde el usuario busca subir de categoría, lo que impulsa la recurrencia.

Además, el vencimiento de puntos:

reduce pasivos financieros
genera picos de consumo (fin de año)

##Código de prueba

```python
from datetime import datetime

from datetime import datetime

class Usuario:
    def __init__(self, id_usuario, nombre):
        self.id = id_usuario
        self.nombre = nombre
        self.visitas = 0
        self.puntos = 0
        self.nivel = "Fan"
        self.historial = []

# Actualizar nivel según visitas
def actualizar_nivel(usuario):
    if usuario.visitas >= 24:
        usuario.nivel = "Super Fanatico"
    elif usuario.visitas >= 12:
        usuario.nivel = "Fanatico"
    else:
        usuario.nivel = "Fan"

# Calcular puntos
def calcular_puntos(usuario, monto):
    if usuario.nivel == "Fan":
        tasa = 0.05
    else:
        tasa = 0.10
    return monto * tasa

# Registrar transacción
def registrar_compra(usuario, monto):
    fecha = datetime.now().date()

    # Validar una visita por día
    if not any(t["fecha"] == fecha for t in usuario.historial):
        usuario.visitas += 1

    # Actualizar nivel
    actualizar_nivel(usuario)

    # Calcular puntos
    puntos_ganados = calcular_puntos(usuario, monto)
    usuario.puntos += puntos_ganados

    # Guardar transacción
    usuario.historial.append({
        "fecha": fecha,
        "monto": monto,
        "puntos": puntos_ganados
    })

    print(f"\nUsuario: {usuario.nombre}")
    print(f"Nivel: {usuario.nivel}")
    print(f"Compra: Q{monto}")
    print(f"Puntos ganados: {puntos_ganados}")
    print(f"Total puntos: {usuario.puntos}")

#Expiración de puntos
def expirar_puntos(usuario):
    usuario.puntos = 0

#Canje de recompensas
def canjear(usuario, costo):
    if usuario.puntos >= costo:
        usuario.puntos -= costo
        print("Canje exitoso")
    else:
        print("Puntos insuficientes")

#Promociones
def aplicar_promocion(dia):
    if dia == "martes":
        return "2x1"


# Crear usuario
usuario = Usuario(1, "Dhidel")

# Simular día
dia = "martes"

# Datos de compra
precio_boleto = 50
cantidad_boletos = 2

# Aplicar promoción
promo = aplicar_promocion(dia)

if promo == "2x1":
    total = precio_boleto  # paga solo uno
else:
    total = precio_boleto * cantidad_boletos

print(f"\nPromoción aplicada: {promo}")
print(f"Total a pagar: Q{total}")

# Registrar compra (esto suma visitas y puntos)
registrar_compra(usuario, total)

# Intentar canjear puntos
print("\nIntentando canje...")
canjear(usuario, 5)

# Mostrar estado final
print("\nEstado final del usuario:")
print(f"Nivel: {usuario.nivel}")
print(f"Puntos: {usuario.puntos}")
print(f"Visitas: {usuario.visitas}")

# Expirar puntos (simulación fin de año)
print("\nExpirando puntos...")
expirar_puntos(usuario)

print(f"Puntos después de expirar: {usuario.puntos}")
```
